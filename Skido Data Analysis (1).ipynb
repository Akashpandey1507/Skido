# â—	Lifetime definition: (Difference in days between ExpireDate and SubscriptionDate)

```python
df['Lifetime'] = (df['ExpireDate'] - df['SubscriptionDate']).dt.days
df['Months'] = df['SubscriptionDate'].dt.month
df['Years'] = df['SubscriptionDate'].dt.year
df['Year-Month'] = df['SubscriptionDate'].dt.to_period('M')
df
```

# â—	Calculate new trials during the month of Jan 2022.

â—	A trial is defined as a user having a lifetime of at most 7 days.

```python
# Filter the Data for January Month
Jan_Trials_Data = df.loc[(df['Years'] == 2022) & (df['Months'] == 1) & (df['Lifetime'] <= 7)]
Jan_Trials_Data
```

```python
Number_of_Trial = Jan_Trials_Data.shape[0]
f"New trials during Jan 2022: {Number_of_Trial}."
```

# â—	Calculate new subscriptions during the month of Jan 2022.

```python
df
```

*â—	A subscription is defined as a user having a lifetime of at least 8 days.

```python
# Calculate new subscriptions during the month of Jan 2022
Jan_Subscriptions = df[(df['Years'] == 2022) & (df['Months'] == 1) & (df['Lifetime']  >= 8)]
Jan_Subscriptions
```

```python
Number_of_Jan_Subscriptions = Jan_Subscriptions.shape[0]
f"New Subscriptions during Jan 2022: {Number_of_Jan_Subscriptions}."
```

# â—	Calculate the trials to subscriptions during the same month. . Express the result as a percentage rounded off to 2 decimal points

â—	Trials to Subscriptions = # Number_of_Jan_Subscriptions / # Number_of_Trial

```python
# Calculate the trials to subscriptions during the same month
Trials_to_Subscriptions = Number_of_Jan_Subscriptions / Number_of_Trial
Trials_to_Subscriptions
```

```python
# Calculate in percentage rounded off to 2 decimal points.
Trials_to_Subscriptions = Trials_to_Subscriptions * 100
f"Trials to Subscriptions during the month of Jan 2022: {round(Trials_to_Subscriptions,2)} %"
```

# â—	Calculate the number of reactivations during the month of Feb 2022

```python
# Calculate the number of reactivations during the month of Feb 2022
Feb_Reactivations = df.loc[(df['Years'] == 2022) & (df['Months'] == 2) & (df['Lifetime'] >= 8)]
Feb_Reactivations
```

```python
Number_of_Feb_Reactivations = Feb_Reactivations.shape[0]
f"Reactivations during Feb 2022: {Number_of_Feb_Reactivations}."
```

# â—	Calculate the retention of subscribers MoM (month over month) and WoW (week over week).

```python
df
```

# â—	Calculate the retention of subscribers MoM (month over month) and WoW (week over week).

```python
# Calculate the retention of subscribers MoM (month over month)
Retention_MOM = df.groupby(df['Year-Month']).size().pct_change().fillna(0) * 100
Retention_MOM
```

```python
plt.figure(figsize=(25,10))
plt.title('Retention of Subscribers Month Over Month',fontsize=25)
plt.xlabel('Retention of Subscriberss',fontsize=16)
plt.ylabel('Distribution of Retention of Subscribers',fontsize=16)


sns.barplot(data=Retention_MOM,x=Retention_MOM.index,y=Retention_MOM.values)
```

```python
df['Week'] = df['SubscriptionDate'].dt.to_period('W')
df
```

```python
# Calculate the retention of subscribers WoW (week over week)
Retention_WOW = round(df.groupby(df['Week']).size().pct_change().fillna(0) * 100,2)
Retention_WOW
```

```python
plt.figure(figsize=(25,10))
plt.title('Retention of Subscribers Weekly',fontsize=25)
plt.xlabel('Retention of Subscriberss',fontsize=16)
plt.ylabel('Distribution of Retention of Subscribers',fontsize=16)
# plt.xticks(rotation=90)


sns.barplot(data=Retention_WOW,x=Retention_WOW.index,y=Retention_WOW.values)
```

# Find Top-5 products for trials MoM

```python
Top_Trial = df.loc[df['Year-Month'] == '2022-01'].groupby('IdProduct').size().nlargest(5)
Top_Trial
```

```python
plt.figure(figsize=(25,10))
plt.title('Top-5 products for trials Month-Over-Month',fontsize=25)
plt.xlabel('Id_Product',fontsize=16)
plt.ylabel('Total Number of Elements in Id_Product',fontsize=16)


sns.barplot(data=Top_Trial,x=Top_Trial.index,y=Top_Trial.values)
```

# Find Top-5 products for subscriptions MoM

```python
Top_Subscriptions = df[df['SubscriptionDate'].dt.to_period('M') == '2022-01'].groupby('IdProduct').size().nlargest(5)
Top_Subscriptions
```

```python
plt.figure(figsize=(25,10))
plt.title('Top-5 products for trials Month-Over-Month',fontsize=25)
plt.xlabel('Id_Product',fontsize=16)
plt.ylabel('Total Number of Elements in Id_Product',fontsize=16)


sns.barplot(data=Top_Subscriptions,x=Top_Subscriptions.index,y=Top_Subscriptions.values)
```
